```python
import pandas as pd
```

```python
df = pd.read_excel('/contracts_plus_loses_dash_0409.xlsx')
```

```python
# Заполняем пропуски в даных
# для Фамилии и имени заменяем на моду
df.middle_name = df.middle_name.fillna((df.middle_name.mode())[0])
df.first_name = df.first_name.fillna((df.middle_name.mode())[0])
# для потерь - id и сумма пропуски заполняем 0 и для имени заполняем как "нет потерь"
df.loss_payout_amt_USD = df.loss_payout_amt_USD.fillna(0)
df.loss_id = df.loss_id.fillna(0)
df.loss_name = df.loss_name.fillna('без потерь')
```

```python
# итого нет пропущенных значений
missing_values_percent = ( (df.isna().sum() / len(df))* 100 ).sort_values(ascending=False)
missing_values_percent
```

```python
# удаляем незначащие для выявления аномалий столбцы
df.drop(['contract_num', 'last_name', 'first_name', 'middle_name', 'loss_id'],  axis=1, inplace = True)
```

```python
# для использования аномалий перекодируем все признаки - в числовые

df_onehot = df.copy()
from sklearn.preprocessing import OneHotEncoder

onehotencoder = OneHotEncoder(sparse = False)
encoded_df = pd.DataFrame(onehotencoder.fit_transform(df_onehot[['product_name']]))
encoded_df.columns = onehotencoder.get_feature_names_out()
df_onehot = df_onehot.join(encoded_df)
df_onehot.drop('product_name', axis = 1, inplace = True)

onehotencoder = OneHotEncoder(sparse = False)
encoded_df = pd.DataFrame(onehotencoder.fit_transform(df_onehot[['contract_status']]))
encoded_df.columns = onehotencoder.get_feature_names_out()
df_onehot = df_onehot.join(encoded_df)
df_onehot.drop('contract_status', axis = 1, inplace = True)

onehotencoder = OneHotEncoder(sparse = False)
encoded_df = pd.DataFrame(onehotencoder.fit_transform(df_onehot[['currency_name']]))
encoded_df.columns = onehotencoder.get_feature_names_out()
df_onehot = df_onehot.join(encoded_df)
df_onehot.drop('currency_name', axis = 1, inplace = True)

onehotencoder = OneHotEncoder(sparse = False)
encoded_df = pd.DataFrame(onehotencoder.fit_transform(df_onehot[['country']]))
encoded_df.columns = onehotencoder.get_feature_names_out()
df_onehot = df_onehot.join(encoded_df)
df_onehot.drop('country', axis = 1, inplace = True)

onehotencoder = OneHotEncoder(sparse = False)
encoded_df = pd.DataFrame(onehotencoder.fit_transform(df_onehot[['sex']]))
encoded_df.columns = onehotencoder.get_feature_names_out()
df_onehot = df_onehot.join(encoded_df)
df_onehot.drop('sex', axis = 1, inplace = True)

onehotencoder = OneHotEncoder(sparse = False)
encoded_df = pd.DataFrame(onehotencoder.fit_transform(df_onehot[['loss_name']]))
encoded_df.columns = onehotencoder.get_feature_names_out()
df_onehot = df_onehot.join(encoded_df)
df_onehot.drop('loss_name', axis = 1, inplace = True)

df_onehot
```

```python
from sklearn.neighbors import LocalOutlierFactor
from sklearn.ensemble import IsolationForest
```

```python
# выявляем аномалии по методу LOF
clf = LocalOutlierFactor(n_neighbors=5)
pred = clf.fit_predict(df_onehot)
```

```python
# для некоторых столбцов разница существенная
df_onehot[pred == -1].mean() - df_onehot.mean()
```

```python
(pred == -1).sum()
```

```python
clf = IsolationForest()
pred_if = clf.fit_predict(df_onehot)
(pred_if == -1).sum()
```

```python
# с помощью метода CLF найдено 556 аномалий и с помощью метода IF 161
df_anomaly = df_onehot[pred == -1]
df_noanomaly = df_onehot[pred != -1]
df_anomaly['anomaly'] = 1
df_noanomaly['anomaly'] = 0
```

```python
# соединяем полный фрейм
df_noanomaly = df_noanomaly.append(df_anomaly)
```

```python
df_noanomaly_join = df_noanomaly[['contract_id','client_id','anomaly']]
df_noanomaly_join
```

```python
df_noanomaly_join.to_excel('contracts_plus_loses_dash_0409_anomaly.xlsx')
```
